```python
#code to draw a bar graph 
def bar_graph(img_name,x_items,color,width,title,ylabel,xlabel,ind,legend,alignment,ymax,y_gap,x_axis_names,rotation):
    l=[]#[0]*len(x_items)
    l.append(plt.bar(ind*3,x_items[0],width/(2*len(x_items)),color=color[0],align=alignment)[0])
    for i in range(1,len(x_items)):
        # p1 = plt.bar(ind*3, win, width/4,color="g",align='center')
        l.append(plt.bar((ind*3)+ width/(2*len(x_items)),x_items[i],width/(2*len(x_items)),color=color[i],align=alignment)[0])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.title(title)
    plt.xticks(ind*3, tuple(x_axis_names),rotation=rotation)
    plt.yticks(np.arange(0, ymax, y_gap))
    plt.legend(tuple(l), legend)
    plt.savefig(img_name)
    plt.show()
#     plt.close()
```

```python
#code to draw a pie chart
def pie_chart(img_name,labels,sizes,explode,color):
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90,colors=color)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.savefig(img_name)
    plt.show()
#     plt.close()
```

## Streangth of Team1 wrt Team2

```python
import math
print delv.columns

#calculating the set of all players participated till date in IPL
def calc_player_list(data):
    players_name=(set(data['batsman'].unique())).union(set(data['non_striker']))
    players_name=players_name.union(set(data['bowler']))
    players_name=players_name.union(set(data['fielder']))
    return players_name
players_name=calc_player_list(delv)
print len(players_name)

#initialization of variables
batsman_performance={}
bowlers_performance={}
career_score={}
strengthOfAwrtB=[]
team1_bat_score=[]
team2_bat_score=[]
team1_ball_score=[]
team2_ball_score=[]
for pl in players_name:
    batsman_performance[pl]=[0]*(8)
    bowlers_performance[pl]=[0]*(7)
    career_score[pl]=[[0.0,0.0],[0.0,0.0]]
#grouping the deliveries dataset by matchid, inning and batsman so that we can get the runs of each batsman in each inning
data = delv.groupby(['match_id','inning','batsman']).sum()

#initialization of weights
w1=10
w2=4
w3=2
w4=5
w5=5

# calculation of batting and bowling score of each team
def calc_score(tp):
    tbat_score=0.0
    tball_score=0.0
    tbat_norm=0.0
    tball_norm=0.0
    for pl in  tp:
        #calculation of batting score of each team
        #u=sqrt(number of matches in which batsman did batting/number of matches played by that batsman)
        u=math.sqrt((1.0*batsman_performance[pl][1])/batsman_performance[pl][0])
        #number of hundreds, fifties and thirties scored by a batsman
        hund=batsman_performance[pl][7]
        fif=batsman_performance[pl][6]
        thir=batsman_performance[pl][5]
        v=(w1*hund)+(w2*fif)+(w3*thir)
        #batting average of batsman
        bat_avg=100.0
        if batsman_performance[pl][3]!=0:
            bat_avg=batsman_performance[pl][2]/batsman_performance[pl][3]
        w=(w4*v)+(w5*bat_avg)
        #updating total batting score
        tbat_score+=u*w
        #updating the max value of career score of the batsman
        career_score[pl][0][0]=max(u*w*1.0,career_score[pl][0][0])
        #updating the current normalized career score of the batmans
        if career_score[pl][0][0]!=0:
            career_score[pl][0][1]=(u*w*1.0)/career_score[pl][0][0]
        else:
            career_score[pl][0][1]=0.0
        
        #bowler
        #u=sqrt(number of matches in which bowler did bowling/number of matches played by that bowler)
        u=math.sqrt((1.0*bowlers_performance[pl][1])/bowlers_performance[pl][0])
        #total matches in which bowler took greater than equal to five wickets, three wickets 
        fv_wkt=bowlers_performance[pl][6]
        thr_wkt=bowlers_performance[pl][5]
        total_wkts=bowlers_performance[pl][2]
        v=w1*fv_wkt + w2*thr_wkt +w2*total_wkts
        #bowling average calculated as (runs given by bowler /wkts taken by bowler)
        bowl_avg=100.0
        if bowlers_performance[pl][2]!=0:
            bowl_avg=((bowlers_performance[pl][4]*1.0)/(bowlers_performance[pl][2])) 
        #bowling economy calculated as (runs given by bowler/overs done by bowler)
        bowl_eco=20
        if bowlers_performance[pl][3]!=0:
            bowl_eco=(bowlers_performance[pl][4]*6.0)/(bowlers_performance[pl][3]) 
        w=bowl_avg*bowl_eco
        #updating the max career score of bowler
        if w>0:
            career_score[pl][1][0]=max((1.0*(u+v))/w,career_score[pl][1][0])
        #updating the total bowling score of the team
        if w>0:
            tball_score+=(u+v)/w
        #updating the current normalised career score of bowler
        if career_score[pl][1][0]!=0:
            career_score[pl][1][1]=((1.0*(u+v))/w)/career_score[pl][1][0]
        else:
            career_score[pl][1][1]=0.0
        
        tbat_norm+=career_score[pl][0][1]
        tball_norm+=career_score[pl][1][1]
#     print tbat_score,tball_score
    return (tbat_norm, tball_norm)


ii=0
#iterating through all the matches
for index,rows in matches.iterrows():
    #colleting all data of ining one and ining2 in seperate dataframes
    d1=delv[(delv['match_id'] ==index+1) & (delv['inning']==1)]
    d2=delv[(delv['match_id'] ==index+1) & (delv['inning']==2)]
    
    #calculating the list of playersin team1 and team2
    p1_name=calc_player_list(d1)
    p2_name=calc_player_list(d2)
    
    #grouping all data by bowler of that match
    dd1=delv[(delv['match_id'] ==index+1)].groupby(['bowler']).sum()
    
    #getting the list of all bowlers in that match
    bowler_name={}
    for name, group in dd1.iterrows():
        #updating the data that the bowler has bowled in this match by incrementing the count by one
        bowlers_performance[name][1]+=1
        bowler_name[name]=0
    
    #iterating through inning 1 dataframe and updating performance of batsman and bowlers in that inning
    for idx,row1 in d1.iterrows():
        #runs hit by batsman in that inning
        batsman_performance[row1['batsman']][2]+=row1['batsman_runs']
        #balls played by batsman till now
        batsman_performance[row1['batsman']][4]+=1
        #updaing if batsman is dissmissed or not
        if row1['player_dismissed']!=0:
            batsman_performance[row1['player_dismissed']][3]+=1
        #updaing if bowler took a wicket
        if row1['dismissal_kind']!=0 and row1['dismissal_kind']!='retired hurt' and row1['dismissal_kind']!='run out':
            bowlers_performance[row1['bowler']][2]+=1
            bowler_name[row1['bowler']]+=1
        #total bowl done by bowler
        bowlers_performance[row1['bowler']][3]+=1
        #total runs given by the bowler in that ball
        bowlers_performance[row1['bowler']][4]+=row1['total_runs']
        
    
    #iterating through inning 2 dataframe and updating performance of batsman and bowlers in that inning
    for idx,row1 in d2.iterrows():
        #runs hit by batsman in that inning
        batsman_performance[row1['batsman']][2]+=row1['batsman_runs']
        #balls played by batsman till now
        batsman_performance[row1['batsman']][4]+=1
        #updaing if batsman is dissmissed or not
        if row1['player_dismissed']!=0:
            batsman_performance[row1['player_dismissed']][3]+=1
        #updaing if bowler took a wicket
        if row1['dismissal_kind']!=0 and row1['dismissal_kind']!='retired hurt' and row1['dismissal_kind']!='run out':
            bowlers_performance[row1['bowler']][2]+=1
            bowler_name[row1['bowler']]+=1
        #total bowl done by bowler
        bowlers_performance[row1['bowler']][3]+=1
        #total runs given by the bowler in that ball
        bowlers_performance[row1['bowler']][4]+=row1['total_runs']
    
    
    
    
    
    
    #list of all players in that match
    p_name=p1_name.union(p2_name)
    #updating that the player has played that match
    for pl in p_name:
        batsman_performance[pl][0]+=1
        bowlers_performance[pl][0]+=1
        
    #groupby batsman so that we can get the total runs of each batsman in that match
    dd1=delv[(delv['match_id'] ==index+1)].groupby(['batsman']).sum()
    for name, group in dd1.iterrows():
        #updating that the player batted in that match
        batsman_performance[name][1]+=1
        #updating >= (100,50,30) runs 
        runs=group['batsman_runs']
        if runs>=100:
            batsman_performance[name][7]+=1
        elif runs>=50:
            batsman_performance[name][6]+=1
        elif runs>=30:
            batsman_performance[name][5]+=1
    
    #updating bowler performance
    for bowler in bowler_name:
        #updating that bowler has bowled in that match
        bowlers_performance[bowler][1]+=1
        wickets=bowler_name[bowler]
        #updating wickets>=(5,3)
        if wickets>=5:
            bowlers_performance[bowler][6]+=1
        elif wickets>=3:
            bowlers_performance[bowler][5]+=1
    
    #list of all team1 players
    team1_player=(set(d1['batsman'].unique())).union(set(d1['non_striker']))
    team1_player=team1_player.union(set(d2['bowler']))
    team1_player=team1_player.union(set(d2['fielder']))
    #calculating score of team1
    team1_score=calc_score(team1_player)
    
    #list of all team2 players
    team2_player=(set(d2['batsman'].unique())).union(set(d2['non_striker']))
    team2_player=team2_player.union(set(d1['bowler']))
    team2_player=team2_player.union(set(d1['fielder']))
    #calculating score of team1
    team2_score=calc_score(team2_player)
    
    #calculation of relative strength
    s_a_b = (team1_score[0]/team2_score[1])-(team2_score[0]/team1_score[1])
    #updating lists
    team1_bat_score.append(team1_score[0])
    team1_ball_score.append(team1_score[1])
    team2_bat_score.append(team2_score[0])
    team2_ball_score.append(team2_score[1])
    strengthOfAwrtB.append(s_a_b)
#     print team1_score,team2_score

#adding new columns to dataframe
matches['SAB']=strengthOfAwrtB
matches['team1_bat_score']=team1_bat_score
matches['team2_bat_score']=team2_bat_score
matches['team1_ball_score']=team1_ball_score
matches['team2_ball_score']=team2_ball_score
```

## Grid Search CV

n_estimators: In general the more the number of trees the less likely the algorithm is to overfit. So we should try to increase this. The lower this number, the closer the model is to a decision tree, with a restricted feature set.

max_features: We should try to reduce this number (something like 30-50% of the number of features). This determines how many features each tree is randomly assigned. The smaller the number, the less likely to overfit, but too small is also not good as it will start to introduce under-fitting.

max_depth: This will reduce the complexity of the learned models, and lowers over-fitting risk. We should start with small, say 5-10, and then increase it to get the best result.

min_samples_leaf: We should set this to values which are greater than one. This has a similar effect to the max_depth parameter, it means the branch will stop splitting once the leaves have that number of samples each

```python
rf_clf=RandomForestClassifier(n_jobs=-1)
n_estm=[]
for i in range(1,20):
    n_estm.append((i+1)*100)
param_grid = { 
    'n_estimators': n_estm,
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,8,10,20,40,60,80,90,100],
    'min_samples_leaf': [1,3,5,7,9,10],
    'min_samples_split': [2,5,10],
    'criterion' :['gini', 'entropy']
}
CV_rf_clf = GridSearchCV(estimator=rf_clf, param_grid=param_grid, cv= 5)
CV_rf_clf.fit(matches[predictor_var], matches['winner'])
```
